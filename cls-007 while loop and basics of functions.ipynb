```python
t  = (3,4,5,6,7,8,9)
```

```python
i = 0
while i < len(t):
    if t[i] == 6:
        print('index of 6 is =', i)
    if t[i] == 7 :
        print('index of 7 is =',i)
    i = i+1
```

```python
i = 0
while i < len(t):
    if t[i] == 6 or t[i]==7:
        print('index of',t[i],' is =', i)
    i = i+1
```
